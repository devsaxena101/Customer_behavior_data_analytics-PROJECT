```python
df['Review Rating'] = df['Review Rating'].fillna(4.1) # fill the null value of review rating is 4.1
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()            # we do the upper case converted into lower case 
df.columns = df.columns.str.replace(' ','_')   # and added _ in which they needed   
df = df.rename(columns=({'purchase_amount_(usd)':'purchase amount'}))  # rename the columns name 
```

```python
df.columns
```

```python
# create a new new columns age_group
labels = ['Young Adult','Adult','Middle aged', 'Senior']   # making a new column age_group
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)    # and devided age  group on the basis of age 
```

```python
df[['age','age_group']].head(20)   # dhows the columns values
```

```python
# create columns purchase_frequency_days
frequency_mapping = {                 
    'Fortnightly': 14,              # replace frequency of rchase columns values with numerical numbers
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Even 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
